# Mini Rag test

```python
from langchain_community.vectorstores import Chroma

texts = [
    "বাংলাদেশ দক্ষিণ এশিয়ার একটি দেশ।",
    "Artificial Intelligence is transforming the world.",
    "ঢাকা বাংলাদেশের রাজধানী।",
    "Machine learning is a subset of AI.",
    "বাংলাদেশের জনসংখ্যা প্রায় ১৭ কোটি।"
]

vectorstore = Chroma.from_texts(
    texts=texts,
    embedding=embeddings
)

# Configure retriever with similarity search and limit results
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 2}  # Return top 2 most similar documents
)

docs = retriever.invoke("বাংলাদেশ কোথায় অবস্থিত?")
print("Retrieved documents:")
for i, doc in enumerate(docs, 1):
    print(f"{i}. {doc.page_content}")

# Let's also test with similarity scores
print("\nWith similarity scores:")
results = vectorstore.similarity_search_with_score("বাংলাদেশ কোথায় অবস্থিত?", k=3)
for doc, score in results:
    print(f"Score: {score:.4f} | Text: {doc.page_content}")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_ollama import OllamaEmbeddings

# Initialize embeddings
embeddings = OllamaEmbeddings(model="mxbai-embed-large")

# Sample texts
texts = [
    "বাংলাদেশ দক্ষিণ এশিয়ায় অবস্থিত।",
    "Artificial Intelligence is transforming the world.",
    "ঢাকা বাংলাদেশের রাজধানী।",
    "Machine learning is a subset of AI.",
    "বাংলাদেশের জনসংখ্যা প্রায় ১৭ কোটি।",
    "Bangladesh is located in South Asia"
]

# Metadata + IDs
metadatas = [{"text": t} for t in texts]
ids = [f"doc{i}" for i in range(len(texts))]

# Create Chroma vectorstore
vectorstore = Chroma.from_texts(
    texts=texts,
    embedding=embeddings,
    metadatas=metadatas,
    ids=ids,
    persist_directory="./chroma_test_db"
)

# Retriever
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)

# Query
query = "বাংলাদেশ কোথায় অবস্থিত?"

# Correct way to fetch documents using invoke (not get_relevant_documents)
docs = retriever.invoke(query)

print("Retrieved documents:")
for i, doc in enumerate(docs, 1):
    print(f"{i}. {doc.page_content}")

# Similarity scores (direct from vectorstore)
results = vectorstore.similarity_search_with_score(query, k=3)
print("\nWith similarity scores:")
for doc, score in results:
    print(f"Score: {score:.4f} | Text: {doc.page_content}")
```
